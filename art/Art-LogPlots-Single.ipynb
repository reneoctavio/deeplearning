```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline

MODEL_PATH = '/Users/reneoctavio/Documents/Paintings/vgg/'

LOSS_FILE = MODEL_PATH + 'log-loss.npy'
ACC1_FILE = MODEL_PATH + 'log-acc1.npy'
ACC5_FILE = MODEL_PATH + 'log-acc5.npy'
LRAT_FILE = MODEL_PATH + 'log-lr.npy'

loss = np.load(LOSS_FILE)
acc1 = np.load(ACC1_FILE)
acc5 = np.load(ACC5_FILE)
lrat = np.load(LRAT_FILE)

print(loss.shape)
plt.plot(loss[:,0], loss[:,1])
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()
```

```python
plt.plot(acc1[:,0], acc1[:,1], label='Top-1')
plt.plot(acc5[:,0], acc5[:,1], label='Top-5')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Score')
plt.show()
```
