```python
import numpy as np
import matplotlib.pyplot as plt

%pylab inline

FT0512_PATH = '/Users/reneoctavio/Documents/leafsnap/finetune_512/'
FT1024_PATH = '/Users/reneoctavio/Documents/leafsnap/finetune_caffe/'
FTGOGN_PATH = '/Users/reneoctavio/Documents/leafsnap/bvlc_googlenet/'

LOSS_FT0512_FILE = FT0512_PATH + 'log-loss.npy'
LOSS_FT1024_FILE = FT1024_PATH + 'log-loss.npy'
LOSS_FTGOGN_FILE = FTGOGN_PATH + 'log-loss.npy'

loss_ft0512 = np.load(LOSS_FT0512_FILE)
loss_ft1024 = np.load(LOSS_FT1024_FILE)
loss_ftgogn = np.load(LOSS_FTGOGN_FILE)

ACC1_FT0512_FILE = FT0512_PATH + 'log-acc1.npy'
ACC1_FT1024_FILE = FT1024_PATH + 'log-acc1.npy'
ACC1_FTGOGN_FILE = FTGOGN_PATH + 'log-acc1.npy'

acc1_ft0512 = np.load(ACC1_FT0512_FILE)
acc1_ft1024 = np.load(ACC1_FT1024_FILE)
acc1_ftgogn = np.load(ACC1_FTGOGN_FILE)

ACC5_FT0512_FILE = FT0512_PATH + 'log-acc5.npy'
ACC5_FT1024_FILE = FT1024_PATH + 'log-acc5.npy'
ACC5_FTGOGN_FILE = FTGOGN_PATH + 'log-acc5.npy'

acc5_ft0512 = np.load(ACC5_FT0512_FILE)
acc5_ft1024 = np.load(ACC5_FT1024_FILE)
acc5_ftgogn = np.load(ACC5_FTGOGN_FILE)

LRAT_FT0512_FILE = FT0512_PATH + 'log-lr.npy'
LRAT_FT1024_FILE = FT1024_PATH + 'log-lr.npy'
LRAT_FTGOGN_FILE = FTGOGN_PATH + 'log-lr.npy'

lrat_ft0512 = np.load(LRAT_FT0512_FILE)
lrat_ft1024 = np.load(LRAT_FT1024_FILE)
lrat_ftgogn = np.load(LRAT_FTGOGN_FILE)

plt.plot(loss_ft0512[:,0], loss_ft0512[:,1], label='Reduced-512')
plt.plot(loss_ft1024[:,0], loss_ft1024[:,1], label='CaffeNet')
plt.plot(loss_ftgogn[:,0], loss_ftgogn[:,1], label='GoogLeNet')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.show()

#OBS: Batch size of GoogLeNet = 32
#     Batch size of others = 256
```

```python
plt.plot(acc1_ft0512[:,0], acc1_ft0512[:,1], label='Reduced-512')
plt.plot(acc1_ft1024[:,0], acc1_ft1024[:,1], label='CaffeNet')
plt.plot(acc1_ftgogn[:,0], acc1_ftgogn[:,1], label='GoogLeNet')
#plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Top-1 Accuracy')
plt.show()
```

```python
plt.plot(acc5_ft0512[:,0], acc5_ft0512[:,1], label='Reduced-512')
plt.plot(acc5_ft1024[:,0], acc5_ft1024[:,1], label='CaffeNet')
plt.plot(acc5_ftgogn[:,0], acc5_ftgogn[:,1], label='GoogLeNet')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Top-5 Accuracy')
plt.show()
```

```python
plt.plot(lrat_ft0512[:,0], lrat_ft0512[:,1], label='Reduced-512')
plt.plot(lrat_ft1024[:,0], lrat_ft1024[:,1], label='CaffeNet')
plt.plot(lrat_ftgogn[:,0], lrat_ftgogn[:,1], label='GoogLeNet')
plt.legend()
plt.xlabel('Iteration')
plt.ylabel('Learning Rate')
plt.show()
```
